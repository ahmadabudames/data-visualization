```python
from matplotlib import pyplot as plt
import seaborn as sns
from calendar import month_abbr
```

```python
sns.load_dataset('planets')
```

# Load the data

```python
sns.get_dataset_names() 
df = sns.load_dataset("planets")
df.head()
```

```python
number = df[df.number >= 1]
distance_by_method = number.groupby("method", as_index=False).distance.mean()
sns.barplot(x="method",y="distance",data=distance_by_method)
distance_by_method
```

# count for the number 

```python
sns.countplot(x="number",data=df)
```

# relplot for the year and the distance

```python
sns.relplot(x="year",y="distance",data=df)
```

# Implot for year and the number  

```python
sns.lmplot(x="year",y="number",data=df)
```

# barplot for year and the mass

```python
sns.barplot(x="year",y="mass", data=df)
```
