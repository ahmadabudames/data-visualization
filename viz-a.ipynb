# Load the data

```python
sns.get_dataset_names() # requires bs4
df = sns.load_dataset("iris")
df.head()
```

# Summarize your insights at bottom of notebook

```python
for_petal_width = df[df.petal_width >= 0.2]
sepal_length_by_method = for_petal_width.groupby("species", as_index=False).sepal_width.mean()
sns.barplot(x="species",y="sepal_width",data=sepal_length_by_method)
sepal_length_by_method
```

# count for the petal_width

```python
sns.countplot(x="petal_width",data=df)
```

# relplot for the species and the sepal_length

```python
sns.relplot(x="species",y="sepal_length",data=df)
```

```python
sns.barplot(x="petal_width",y="sepal_width", data=df)
```
